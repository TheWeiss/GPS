# Move Best Models to production:

```python
import pandas as pd
from tqdm import tqdm
import os
import glob
from distutils.dir_util import copy_tree
from distutils.file_util import copy_file
```

```python
results = pd.read_csv('res_summery.csv')
results = results[['species', 'antibiotic_name', 'exp_path']]
results['antibiotic_name'] = results['antibiotic_name'].str.lower()
results['antibiotic_name'] = results['antibiotic_name'].replace(' ', '_', regex=True)
results['antibiotic_name'] = results['antibiotic_name'].replace('-', '_', regex=True)
results.head().values
```

```python
valid_h2o_paths = []
valid_autixgb_paths = []
for exp_dir in tqdm(os.listdir('experiments')):
    try:
        model_dir = glob.glob('experiments/' + exp_dir + '/model_h2o*')[0]
        valid_h2o_paths.append(exp_dir)
    except:
        print(exp_dir + ': is missing h2o model')
        
    try:
        model_dir = glob.glob('experiments/' + exp_dir + '/model_autoxgb*')[0]
        valid_autixgb_paths.append(exp_dir)
    except:
        print(exp_dir + ': is missing autoxgb model')
        
```

```python
valid_h2o_paths
```

```python
def parse_model_path(model_dir):
    print(model_dir)
    model_param = {}
    model_param['model_name'] = model_dir.split('_')[1]
    if model_param['model_name'] == 'autoxgb':
        model_param['num_models'] = float(model_dir.split('_')[4])
        model_param['time_limit'] = float(model_dir.split('_')[7])
    elif model_param['model_name'] == 'h2o':
        model_param['num_models'] = float(model_dir.split('_')[7])
        model_param['time_limit'] = float(model_dir.split('_')[4])
    return model_param
```

```python
for exp_path in valid_h2o_paths:
    data_param = pd.read_csv('experiments/{}/data_param.csv'.format(exp_path))
    species = data_param['species'].iloc[0]
    antibiotic = data_param['antibiotic'].iloc[0]
    antibiotic = antibiotic.replace('-', '_')
    antibiotic = antibiotic.replace(' ', '_')
    os.makedirs('models/{}'.format(species), exist_ok=True)
    os.makedirs('models/{}/{}'.format(species, antibiotic), exist_ok=True)
    model_dir = glob.glob('experiments/{}/model_h2o*'.format(exp_path))[-1]
    print(model_dir)
    model_param = parse_model_path(model_dir.split('/')[-1])
    print(model_param)
    copy_tree('{}/model'.format(model_dir), 'models/{}/{}/model'.format(species, antibiotic))
    copy_file('experiments/{}/data_param.csv'.format(exp_path), 'models/{}/{}/data_param.csv'.format(species, antibiotic))
    copy_file('experiments/{}/features.csv'.format(exp_path), 'models/{}/{}/features.csv'.format(species, antibiotic))
    pd.DataFrame(model_param, index=[0]).to_csv('models/{}/{}/model_param.csv'.format(species, antibiotic))
```
